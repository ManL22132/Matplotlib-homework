```python
%matplotlib inline
# Dependencies and Setup

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"
```

```python
# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)


# Combine the data into a single dataset
city_ride_data_complete = pd.merge(city_data, ride_data, how="left", on=["city", "city"])

# Display the data table for preview
city_ride_data_complete
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
city_data_df = city_ride_data_complete.groupby(["city", "type"])
city_data_df

city_cnt = city_data_df["city"].count()
type_cnt = city_data_df["type"].count()
total_fare = city_data_df["fare"].sum()
avg_fare = total_fare / city_cnt 

# Build the scatter plots for each city types
# create data
x = city_cnt
y = avg_fare
z = x
z = z*z

plt.scatter(city_cnt, avg_fare, s=z*1000, c=x,  alpha=0.4, edgecolors="grey", linewidth=2)


# Incorporate the other graph properties

# Create a legend
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.title("Pyber Ride Sharing Data (2016)")
plt.grid(b=None, which='major', axis='both')
plt.legend(loc='upper right')

# Incorporate a text label regarding circle size

# Save Figure
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
city_data_df = city_ride_data_complete.groupby(["type"])
total_fare = city_data_df["fare"].sum()
total_fare

# Build Pie Chart
# Data to plot
labels = 'Rural', 'Suburban' ,'Urban' 
sizes = total_fare
colors = ['gold', 'lightskyblue', 'lightcoral']
explode = (0, 0, 0.1)  # explode 1st slice
 
# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=150)

plt.title("% of Total Fares by City Type")

plt.axis('equal')

# Show Figure
plt.show()
```
